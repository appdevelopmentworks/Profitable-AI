乳がん診断データセットの例


```python
import warnings
import pandas as pd
import numpy as np
#
import matplotlib.pyplot as plt
#
import seaborn as sns
#グラフ日本語表示用のライブラリ
import japanize_matplotlib
#データフレームを見やすく表示するライブラリ
from IPython.display import display
#乳がんのデータセット
from sklearn.datasets import load_breast_cancer
#訓練データとテストデータの分割ライブラリ
from sklearn.model_selection import train_test_split
#ロジステック回帰モデル
from sklearn.linear_model import LogisticRegression
```

```python
#警告は出さない
warnings.filterwarnings('ignore')
```

```python
#numpyをプリントで表示するときの設定、指数表示あり、少数点以下の桁数４
np.set_printoptions(suppress=True, precision=4)
#データフレーム内のfloat型は小数点４桁までの表示
pd.options.display.float_format = '{:.4f}'.format
#データフレームの表示最大行数設定
pd.set_option("display.max_columns", None)
#plotのフォントサイズをデフォルト指定
plt.rcParams["font.size"] = 14
#乱数の種を固定
random_seed = 123
```

```python
#乳がんデータを読み込み
cancer = load_breast_cancer()
#注釈を表示
print(cancer.DESCR)
```

```python
#日本語の列名をリスト
columns = [
    '半径_平均', 'きめ_平均', '周長_平均', '面積_平均', 
    '平滑度_平均','コンパクト度_平均', '凹面_平均',
    '凹点_平均', '対称性_平均', 'フラクタル度_平均',
    '半径_標準誤差', 'きめ_標準誤差', '周長_標準誤差',
    '面積_標準誤差', '平滑度_標準誤差',
    'コンパクト度_標準誤差', '凹面_標準誤差', '凹点_標準誤差',
    '対称性_標準誤差', 'フラクタル度_標準誤差',
    '半径_最大', 'きめ_最大', '周長_最大', '面積_最大', 
    '平滑度_最大','コンパクト度_最大', '凹面_最大', '凹点_最大', 
    '対称性_最大', 'フラクタル度_最大'
]

#データセットの説明変数と、日本語列名でデータフレーム作成
df = pd.DataFrame(cancer.data, columns=columns)
#目的変数をシリーズとして
y = pd.Series(cancer.target)
```

```python
df
```

```python
#説明変数の確認
#インデックス20ー25を表示
display(df[20:25])
```

```python
#目的変数の確認インデックス20ー25
print(y[20:25])
```

```python
#データセットのサイズ
print(df.shape)

#説明変数の要素ごとのカウント値
print(y.value_counts())
```

```python
#
#

#０（悪性）のみのデータセット
df0 = df[y==0]

#１（良性）のみのデータセット
df1 = df[y==1]
#2つに分けたデータセットを並べて表示
display(df0.head())
display(df1.head())
```

```python
dfh = pd.DataFrame(df, columns = df.columns).assign(判定=np.array(y))
dfh.head(3)
```

```python
#ヒートマップで可視化
#各列間の相関係数を算出
heat_map = dfh.corr(method="pearson")
#ヒートマップ設定
sns.heatmap(heat_map, center=0, annot=True, fmt="1.1f")
#ヒートマップとデータセットを表示
plt.show()
#データフレーム内小数点1桁表示に
pd.options.display.float_format = '{:.1f}'.format
display(heat_map)
```

```python
df.columns
```

```python
#
#グラフの表示サイズ
plt.figure(figsize=(6, 6))

#散布図（悪性）のplot
plt.scatter(df0['半径_平均'], df0['きめ_平均'], marker='x', c='b', label='悪性')

#散布図（良性）のplot
plt.scatter(df1['半径_平均'], df1['きめ_平均'], marker='s', c='k', label='良性')
#グリッド表示
plt.grid()
#凡例を設定
plt.xlabel('半径_平均')
plt.ylabel('きめ_平均')
#凡例を表示
plt.legend()
#全体表示
plt.show()
```

```python
#特徴量を2つに絞る
input_columns = ['半径_平均', 'きめ_平均']
#特徴量が半径_平均、きめ_平均のみのデータセット
x = df[input_columns]
display(x.head())
```

```python
#訓練データとテストデータに分割
x_train, x_test, y_train, y_test =train_test_split(x, y, train_size=0.7, test_size=0.3, random_state=random_seed)
```

```python
#それぞれのサイズ確認
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#それぞれのデータセットを並べて表示
display(x_train.head())
display(x_test.head())
display(y_train.head())
display(y_test.head())
```

```python
#アルゴリズムにロジスティック回帰を使用
algorithm = LogisticRegression(random_state=random_seed)
```

```python
#学習
algorithm.fit(x_train, y_train)
print(algorithm)
```

```python
#テストデータの説明変数から予測させる
y_pred = algorithm.predict(x_test)
#予測値確認
y_pred
```

```python
#正解データ（テストデータ）と予測したデータを並べて表示
#8件目の予測が外れている！
#正解データ先頭から10個
y_test10  = y_test[:10].values
print(y_test10)
#予測データ先頭から10個
y_pred10 = y_pred[:10]
print(y_pred10)
```

```python
#
#正解がTrueで間違いがFalseで格納して表示
w1 = (y_test10==y_pred10)
print(w1)
#正解の数を合計して表示
w2 = w1.sum()
print(w2)
```

```python
#テストデータの正解すべてとテストデータから予測したデータ全て
#正解がTrueで間違いがFalseでｗに格納
w = (y_test.values==y_pred)
#正解の合計
correct = w.sum()
#w配列の長さ（データ数）
N = len(w)
#正解率
score = correct / N
#小数点以下4桁で表示
print(f'精度：{score:.04f}')
```

```python
#
#scoreメソドを使えば簡単に精度を求められる
score = algorithm.score(x_test, y_test)
print(f'精度：{score:.04f}')
```

```python
#説明変数を全て使って、訓練データとテストデータを作成
x2_train, x2_test, y_train, y_test =train_test_split(df, y, train_size=0.7, test_size=0.3, random_state=random_seed)
#アルゴリズムはロジステック回帰
algorithm2 = LogisticRegression(random_state=random_seed)
#学習
algorithm2.fit(x2_train, y_train)
#モデルの精度を算出
score2 = algorithm2.score(x2_test, y_test)
#表示、先ほど（説明変数2つ）より精度は上がる
print(f'精度：{score2:.04f}')
```
